## groupby() function

Used to split the data into groups based on some criteria.
When groupby function is performed, it involves one of the following operations on the original object:<br>
&emsp; 1. Spliting the Object<br>
&emsp; 2. Applying a function<br>
&emsp; 3. Combining the result

```python
df=pd.read_csv("E:\\Machine Learning\\student_results_2.csv",index_col='ID')
display(df)
```

```python
gr1=df.groupby(by="StudyHrs") # grouping the values using StudyHrs column.
display(gr1)
display(gr1.groups)
gr2=df.groupby(by=["Class","StudyHrs"]) # grouping the values using Class and StudyHrs columns.(multiple columns used for grouping)
display(gr2.groups)
for class_num,df1 in gr2:
    display(class_num)
    display(df1)
    display()
```

```python
# to find out some special group we can use get_group() method with the group object.
df3=gr1.get_group(3)
display(df3)
```

```python
display(gr1.sum()) # used to get the sum of all the columns according to the groups formed
display(gr1.mean()) # used to get the mean of all the columns according to the groups formed
display(gr1.describe()) # used to get the summary of the dataframe
```

```python
display(gr1.agg(['sum','max','mean'])) # if we want to use multiple functions then we have to give all the function names in a list passed to aggregate function.
```

## merge() function

Used to merge multiple dataframe based on one or more keys.

```python
df1=pd.DataFrame({'ID':range(1,5),'Class':range(9,13)})
display(df1)
df2=pd.DataFrame({'ID':[1,2,3,5],'Name':['A','B','A','C']})
display(df2)
```

```python
# merging this two dataframes using merge function.
display(pd.merge(df1,df2,on="ID"))
# if we try to merge dataframes in which some indices of column specified in on parameter are available in only one dataframe
# then it will discard those values and will merge only if the value is available in both the dataframes.

# how parameter : default value is inner (means intersection)(takes value if it is in both dataframes)
# if we want all the rows of first dataframe then we can use how parameter with value 'left'
display(pd.merge(df1,df2,on="ID",how="left"))
# similarly if we want all rows from second dataframe then we use 'right' value
display(pd.merge(df1,df2,on="ID",how="right"))
# and if we want all rows from both the dataframes then we use 'outer' value
display(pd.merge(df1,df2,on="ID",how="outer"))
# after that if we want to know which value belongs to which dataframe then we can use indicator parameter with value 'True'
display(pd.merge(df1,df2,on="ID",how="outer",indicator=True))
```

```python
# suppose we are having two dataframes with same values then it will generate output as shown below
df1=pd.DataFrame({'ID':range(1,5),'Class':range(9,13)})
display(df1)
df2=pd.DataFrame({'ID':range(1,5),'Class':range(9,13)})
display(df2)
display(pd.merge(df1,df2,on="ID"))
# as we can see it generates output like Class_x and Class_y columns as all the values are same.
# if we want to give different suffixes then we can use suffixes parameter with tuple of suffixes.
display(pd.merge(df1,df2,on="ID",suffixes=('_First_DataFrame','_Second_DataFrame')))
```

## concat() function

Used to concat dataframes or Series.

```python
s1=pd.Series(range(1,6))
s2=pd.Series(range(6,10))
display(pd.concat([s1,s2]))
```

```python
df1=pd.DataFrame({'ID':[1,2,3],'Class':['A','B','C']})
df2=pd.DataFrame({'ID':[4,5],'Class':['B','C']})
display(pd.concat([df1,df2],ignore_index=True)) # merging two dataframes row by row
display(pd.concat([df1,df2],axis=1)) # merging two dataframes column by column
display(pd.concat([df1,df2],ignore_index=True)) # merging two dataframes row by row and using ignore_index parameter to have proper indices for each row.
```

```python
# join parameter : default value is 'outer' means union. another value can be 'innner'
display(pd.concat([df1,df2],join='inner'))
display(pd.concat([df1,df2],join='inner',ignore_index=True)) # giving proper indeices to the rows
display(pd.concat([df1,df2],join='inner',axis=1)) # combining dataframes column by column

# keys parameter : default value is 'None'. Used to concat and separate them giving names to the dataframes.
display(pd.concat([df1,df2],keys=['DataFrame 1','DataFrame 2']))
# we can also combine dataframes column by column and then we can separate them using keys parameter.
display(pd.concat([df1,df2],keys=['DataFrame 1','DataFrame 2'],axis=1))
```

```python
# join method : if we are having two dataframes having totally different indeices then we can use join method to concat them effectively
df1=pd.DataFrame({'A':[1,2,3],'B':[10,20,30]})
df2=pd.DataFrame({'C':[4,5,6],'D':[40,50,60]})
display(df1,df2)
display(df1.join(df2))
df3=pd.DataFrame({'E':[7,8],'F':[70,80]})
display(df1.join(df2).join(df3,how="left"))
# if we are having same column names in dataframes then we can use lsuffix or rsuffix parameters
df4=pd.DataFrame({'A':[7,8,9],'C':[70,80,90]})
display(df1.join(df4,lsuffix="_1")) # adding suffix '_1' for all the columns in first dataframe which are present in both dataframes (here 'A')
# similarly we can also give name to the second dataframe columns using rsuffix parameter
display(df1.join(df4,rsuffix="_2"))
# Also we can specify both parameters simultaneously
display(df1.join(df4,lsuffix="_1",rsuffix="_2"))
```

```python
# append() function : as the name suggests it is used to append one dataframe at the end of another.
display(df1.append(df4,ignore_index=True))
```
